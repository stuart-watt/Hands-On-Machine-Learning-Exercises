# Support Vector Machines

In this notebook we take a look into Support Vector Machines (SVMs), which are one of the most popular machine learning models.

## Linear SVM Classification

Linear SVMs work by creating linear desicion boundaries (similar to logistic regression) where the boundaries are those that are as far as possible from the closest training instances to that boundary. This is called *large margin classification*. The boundary is then characterised by the closest training instances, so called the *support vectors*.

One important thing to note is that SVMs are sensitive to feature scales, so scaling is essential here.

The classification can then be *soft* or *hard*. In hard margin classification, all the instances must be off the *street* (the area around the decision boundary) and on the correct side. This only works if the data is linearly separable and then it is sensitive to outliers. In soft margin classification, we allow the street to be flexible to obtain a balance between having the street as wide as possible and limiting the *margin violations* (instances on or across the street). The hardness is controlled by the hyperparameter C (high C is more hard).

Lowering C can help reduce overfitting.

Lets use a linear SVM classifier to classifier the Iris-Virginica flowers in the Iris dataset:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)] # petal length and width
y = (iris["target"] == 2).astype(np.float64) # Iris-Virginica

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_scv", LinearSVC(C=1, loss="hinge")),
])

svm_clf.fit(X, y)
svm_clf.predict([[5.5, 1.7]])
```

This can also be done using SVC(kernel="linear", C=1) or even SGDClassifier(loss="hinge, alpha=1/(mC)). However, each of these are slower than the LinearSVC class. 

## Nonlinear SVM Classification

Some datasets are not linearly separable. One way around this is to add more features (such as polynomial features) which then may create a linearly separable set. Lets see this on the Moons dataset:

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures

X, y = make_moons(noise=0.1)

poly_svm_clf = Pipeline([
    ("poly_features", PolynomialFeatures(degree=3)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC(C=10, loss="hinge"))
])

poly_svm_clf.fit(X, y)
```

```python
# create a mesh to cover the data space
mesh_x1, mesh_x2 = np.meshgrid(np.arange(X[:, 0].min() - 0.2, X[:, 0].max() + 0.2, 0.01),
                              np.arange(X[:, 1].min() - 0.2, X[:, 1].max() + 0.2, 0.01))

# use the model to make the predictions on the mesh then plot as a contour
mesh_predict = poly_svm_clf.predict(np.c_[mesh_x1.ravel(), mesh_x2.ravel()])
mesh_predict = mesh_predict.reshape(mesh_x1.shape)

# Plot the actual training data
plt.contourf(mesh_x1, mesh_x2, mesh_predict, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
plt.show()
```

The problem with this method is that for low degree polynomials, the model cannot handle complex datasets, while higher degree polynomials, there are lots of featrues, making the model and training the model slow.

This can be overcome using the *kernel* trick. This trick doesn't actually add features and can be used for higher degrees.

```python
from sklearn.svm import SVC

poly_kernel_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5)) 
    # coef0 determines the influence of higher degree polynomials over lower degree to the model
])
poly_kernel_svm_clf.fit(X, y)

# use the model to make the predictions on the mesh then plot as a contour
mesh_predict = poly_kernel_svm_clf.predict(np.c_[mesh_x1.ravel(), mesh_x2.ravel()])
mesh_predict = mesh_predict.reshape(mesh_x1.shape)

# Plot the actual training data
plt.contourf(mesh_x1, mesh_x2, mesh_predict, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
plt.show()
```

### Similarity features

Another way to handle nonlinearity is to add features computed using a *similarity function*. This function measures how much each instance resembles a *landmark* ( a new "instance" added to the dataset). These new features will depend on the landmarks and the similarity function chosen. Use these to complete the new features for each instance. 

A common similarity function the Gaussian *Radial Bias Function* (RBF).

Selecting the landmarks can be done simply by allowing every instance to be a landmark. Then for m instances, there would be m new features. This is computationally expensive to compute and also to train. So again we can use the kernel trick:

```python
models = (
    Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
]),
    Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=0.1, C=0.001))
]),
    Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=10, C=1000))
]),
    Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=0.1, C=1000))
])
)

titles = (
    "gamma=5, C=0.001",
    "gamma=0.1, C=0.001",
    "gamma=10, C=1000",
    "gamma=0.1, C=1000"
)

models = (clf.fit(X, y) for clf in models)

fig, sub = plt.subplots(2, 2, figsize=(8, 8))
plt.subplots_adjust(wspace=0.1, hspace=0.1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    mesh_predict = clf.predict(np.c_[mesh_x1.ravel(), mesh_x2.ravel()]).reshape(mesh_x1.shape)
    ax.contourf(mesh_x1, mesh_x2, mesh_predict, cmap=plt.cm.coolwarm, alpha=0.8)
    
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(mesh_x1.min(), mesh_x1.max())
    ax.set_ylim(mesh_x2.min(), mesh_x2.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

Here the gamma hyperparameter determines the linewidth of the gaussian bell shape. By making it lower, the bell shape gets wider, so the region of influence of each instances gets large, then the boundary becomes smoother. By increasing gamma, each instance has a smaller influence range and the boundary gets more jumpy (bottom left). So both gamma and C act as regularisation hyperparameters.

While there are lots of kernels available, the most commonly used are linear (so try LinearSVC first) and RBF (try second). The kernel trick (sequential minimal optimisation), has complexity of O(m^3xn), so does not scale well with large datasets. It performs really well on complex datasets however.

## SVM Regression

Instead of trying to fit the widest street that has no instances on it (classification), we can try to fit as many features *on* the street and use the street as a regression model. The street width is controlled with the hyperparameter epsilon, e. These models are *e-insensitive* since adding more instances inside the street does not change the model.

SVM regression can be implemeted using LinearSVR (for linear regression) or SVR (select the various kernels you want). 

SMVs can also be used for outlier detection.
