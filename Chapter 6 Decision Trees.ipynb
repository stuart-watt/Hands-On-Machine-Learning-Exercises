# Chapter 6 Decision Trees

In this chapter we will look at how to build and train decision trees, which are some of the most versitile machine learning models and also the basis for random forest models. Decision trees can perform both regression and classification, just like SVMs.

A decision tree is a *whitebox* model in that it is easy to see how predictions are made and how features are used to make those predictions. Scikit-Learn uses the CART (*Classification and Regression Tree*) algorithm for training decision trees. This type of decision tree is a binary tree, each non-leaf node has exactly two children (essentially asking a yes/no question).

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

The CART algorithm splits the dataset along single features in such a way that creates the *purest* subsets, weighted by the subset size. It will keep splitting the subsets until it reaches the max_depth hyperparameter or it cannot reduce the impurity any more.

The CART algorithm is a *greedy* algorithm. It will look for the best split at the time, even if this split is not the best down the track. So the algorithm may not find the optimum split, it just does the best it can at each level without considering the following levels.

Decision tree models are *nonparametric*, meaning that they do not have a predefined set of parameters. The model structure is free to adapt to the training data. Contrast this to a *parametric* model like a lienar model which has a predetermined list of parameters. This constrains the model which reduces overfitting but may lead to underfitting.

Since decision trees are prone to overfitting, they need to have their freedom limited -- this is done by regularisation. One major regularisation hyperparameter is the max_depth which limits the tree depth.

Another method is to let an unconstrained model train to the dataset, then *prune* off the nodes that are not *statistically significant*.

Lets have a look at the moons dataset:

```python
from sklearn.datasets import make_moons
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

models = (DecisionTreeClassifier(),
          DecisionTreeClassifier(min_samples_leaf=4))

X, y = make_moons(noise=0.2)

# create a mesh to cover the data space
mesh_x1, mesh_x2 = np.meshgrid(np.arange(X[:, 0].min() - 0.2, X[:, 0].max() + 0.2, 0.01),
                              np.arange(X[:, 1].min() - 0.2, X[:, 1].max() + 0.2, 0.01))

titles = (
    "Unconstrained",
    "Constrained"
)

models = (clf.fit(X, y) for clf in models)

fig, sub = plt.subplots(2, 1, figsize=(8, 8))
plt.subplots_adjust(wspace=0.1, hspace=0.1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    mesh_predict = clf.predict(np.c_[mesh_x1.ravel(), mesh_x2.ravel()]).reshape(mesh_x1.shape)
    ax.contourf(mesh_x1, mesh_x2, mesh_predict, cmap=plt.cm.coolwarm, alpha=0.8)
    
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(mesh_x1.min(), mesh_x1.max())
    ax.set_ylim(mesh_x2.min(), mesh_x2.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

The unconstrained model is clearly overfitting, while the constrained model will generalise better.

### Regression

Decision trees can also be used for regression tasks. The model works the same way except that each leaf node returns a value instead of a class. The value returned is the mean value of all the instances associated with that node.

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg_2 = DecisionTreeRegressor(max_depth=4)

X = np.linspace(-10, 10, 100)
y = 0.5*X**2 + X + 6*np.random.randn(100)

tree_reg.fit(X.reshape(-1, 1), y)
tree_reg_2.fit(X.reshape(-1, 1), y)

y_predict = tree_reg.predict(X.reshape(-1, 1))
y_predict_2 = tree_reg_2.predict(X.reshape(-1, 1))

plt.scatter(X, y)
plt.plot(X, y_predict, "k", label="max_depth=2")
plt.scatter(X, y)
plt.plot(X, y_predict_2, "b", label="max_depth=4")
plt.legend()
plt.show()
```

Each leaf node returns a mean value and also has a Mean Squared Error for the instances in that node. The CART algorithm makes the splits that minimises the MSE.

The regression decision trees also need regularisation to prevent overfitting.

One downside to decision trees is their instability. Since the splits are done orthogonally in the feature space, the models are sensitive to data rotations. There are also sensitive to small variations in the dataset such as adding or removing instances. Also, shince the features chosen to split are taken randomly, the models may differ even for the same training set. Setting the random_state hyperparameter can stop this.

Random forests will reduce this instability be training many trees and averaging the results over all of them.
